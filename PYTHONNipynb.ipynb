by deafult user input is stored in string because it is universal data format to reduce. errors

The `input()` function always returns a string for a few key reasons:

1.  **Universality:** Text is the most universal data type. Whatever the user types (numbers, words, symbols), it can always be represented as a string. If it were to guess the type (e.g., int if digits were entered), it would fail for mixed input or even simple text.
2.  **Explicit Conversion:** By returning a string, Python forces the programmer to explicitly convert the input to the desired data type (e.g., `int()`, `float()`, `bool()`). This makes the code more robust and prevents unexpected errors from implicit type conversions. It gives the programmer control over how the data is interpreted.
3.  **Preventing Errors:** If `input()` tried to automatically convert input, it would lead to frequent errors when the input doesn't match the expected type. For example, if a user types "hello" when a number is expected, an automatic conversion would crash the program, whereas returning a string allows the programmer to handle such cases gracefully.

This design choice promotes clearer, safer, and more flexible code.

### Why?

The `~` operator is the **Bitwise NOT (Complement)** operator.

It flips every bit of the number.

#### Step 1: Binary representation of 3

```
3 = 00000011
```

#### Step 2: Flip all bits

```
00000011
↓
11111100
```

In Python, integers use **two's complement** representation for signed numbers.

The result becomes:

```
11111100 = -4
```

### Shortcut Formula

For any integer `x`:

[
\boxed{\sim x = -(x+1)}
]

So,

```
~3
= -(3 + 1)
= -4
```

# Left Shift (`<<`) and Right Shift (`>>`) Operators in Python

These are **bitwise shift operators**. They shift the bits of a number to the left or right.

---

# 1. Left Shift Operator (`<<`)

### Definition

The **left shift (`<<`)** operator shifts all the bits of a number to the **left** by the specified number of positions.

Each left shift by **1** is equivalent to **multiplying by 2**.

### Syntax

```python
number << n
```

where:

* `number` = value to shift
* `n` = number of positions

---

### Example 1

```python
print(5 << 1)
```

**Output**

```text
10
```

### Explanation

Binary of 5:

```text
5 = 00000101
```

Shift left by 1:

```text
00000101
↓
00001010
```

Binary `00001010` = **10**

---

### Example 2

```python
print(5 << 2)
```

Output:

```text
20
```

Explanation:

```text
5 = 00000101

Shift by 2

00010100
```

Binary `00010100` = **20**

---

## Shortcut

```text
x << n = x × 2ⁿ
```

Examples:

| Expression | Result |
| ---------- | ------ |
| `5 << 1`   | 10     |
| `5 << 2`   | 20     |
| `8 << 1`   | 16     |
| `3 << 3`   | 24     |

---

# 2. Right Shift Operator (`>>`)

### Definition

The **right shift (`>>`)** operator shifts all the bits of a number to the **right**.

Each right shift by **1** is equivalent to **integer division by 2**.

### Syntax

```python
number >> n
```

---

### Example 1

```python
print(20 >> 1)
```

Output

```text
10
```

Explanation

```text
20 = 00010100

Shift right by 1

00001010
```

Binary `00001010` = **10**

---

### Example 2

```python
print(20 >> 2)
```

Output

```text
5
```

Explanation

```text
20 = 00010100

Shift right by 2

00000101
```

Binary `00000101` = **5**

---

## Shortcut

```text
x >> n = x // 2ⁿ
```

Examples:

| Expression | Result |
| ---------- | ------ |
| `20 >> 1`  | 10     |
| `20 >> 2`  | 5      |
| `16 >> 3`  | 2      |
| `32 >> 4`  | 2      |

# Difference Between `&` and `and` in Python

| `&` (Bitwise AND)                                          | `and` (Logical AND)                      |
| ---------------------------------------------------------- | ---------------------------------------- |
| Operates on **bits**                                       | Operates on **Boolean expressions**      |
| Compares binary digits                                     | Compares True/False values               |
| Used in bit manipulation                                   | Used in conditions (`if`, `while`)       |
| Evaluates **both operands**                                | Uses **short-circuit evaluation**        |
| Returns an integer (or element-wise result for some types) | Returns a Boolean or one of the operands |

---

# 1. `&` (Bitwise AND)

It compares each bit of two numbers.

### Rule

```text
1 & 1 = 1
1 & 0 = 0
0 & 1 = 0
0 & 0 = 0
```

### Example

```python
a = 5
b = 3

print(a & b)
```

### Binary

```text
5 = 0101
3 = 0011
-------------
    0001
```

Output

```text
1
```

---

# 2. `and` (Logical AND)

Used to combine two conditions.

Returns `True` only if **both conditions are True**.

### Example

```python
x = 10

print(x > 5 and x < 20)
```

Output

```text
True
```

Explanation

```text
10 > 5   → True
10 < 20  → True

True and True = True
```

---

# Another Example

```python
x = 10

print(x > 20 and x < 30)
```

Output

```text
False
```

---

# Short-Circuit Difference

### `and` stops early

```python
a = 5

print(a < 2 and 10 / 0)
```

Output

```text
False
```

No error occurs because the first condition is `False`, so Python doesn't evaluate `10 / 0`.

---

### `&` evaluates both operands

```python
a = 5

print((a < 2) & (10 / 0))
```

Output

```text
ZeroDivisionError
```

Here, `&` evaluates **both** sides, so `10 / 0` is executed.

---

# Example with Booleans

```python
print(True & False)
```

Output

```text
False
```

```python
print(True and False)
```

Output

```text
False
```

Although the outputs are the same here, the purpose is different:

* `&` performs a **bitwise operation** (on booleans, `True` is treated as `1` and `False` as `0`).
* `and` performs a **logical operation**.

---

# Interview Answer (30 Seconds)

> **`&` is the bitwise AND operator. It compares the binary bits of two operands and always evaluates both sides. It is mainly used in bit manipulation. `and` is the logical AND operator used to combine Boolean expressions. It returns `True` only if both conditions are true and uses short-circuit evaluation, meaning it stops evaluating as soon as the result is known.**

### Easy Trick to Remember

* **`&` → Bitwise → Binary Numbers**
* **`and` → Logical → Conditions (`if`, `while`)**

MODULES IN PYTHON
# Modules in Python

## Definition

A **module** is a file containing Python code (functions, variables, classes, etc.) that can be reused in another Python program.

A module helps organize code and avoids writing the same code repeatedly.

---

# Why Do We Use Modules?

* ✅ Code Reusability
* ✅ Better Code Organization
* ✅ Easy Maintenance
* ✅ Avoids Code Duplication

---

# Types of Modules

## 1. Built-in Modules

These are already provided by Python.

Examples:

* `math`
* `random`
* `os`
* `time`
* `datetime`
* `statistics`

Example:

```python
import math

print(math.sqrt(25))
```

**Output**

```
5.0
```

---

## 2. User-Defined Modules

You can create your own module.

### Step 1: Create a file `calculator.py`

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b
```

### Step 2: Import it into another file

```python
import calculator

print(calculator.add(10, 20))
print(calculator.sub(20, 10))
```

Output

```
30
10
```

---

# Ways to Import Modules

## 1. Import Entire Module

```python
import math

print(math.sqrt(16))
```

---

## 2. Import Specific Function

```python
from math import sqrt

print(sqrt(25))
```

---

## 3. Import Multiple Functions

```python
from math import sqrt, factorial

print(sqrt(49))
print(factorial(5))
```

---

## 4. Import Everything

```python
from math import *

print(sqrt(81))
```

⚠️ **Not recommended** because it can create naming conflicts.

---

## 5. Import with Alias

```python
import math as m

print(m.pi)
print(m.sqrt(64))
```
